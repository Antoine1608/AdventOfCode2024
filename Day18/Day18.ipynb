```python
import numpy as np
import pandas as pd
from dijkstar import Graph, find_path
from tqdm import tqdm
```

```python
with open("Input.txt") as file:
    input = file.read().split('\n')
    file.close

df = pd.DataFrame(index = range(71), columns = range(71))
```

```python
for tup in input[0:1024]:
    tup_ = tup.split(',')
    df.iloc[int(tup_[1]), int(tup_[0])] = '#'
        
df = df.fillna('.')
```

```python
# Création d'une marge autour du cadre
df[len(df)] = '#'
df = pd.concat([df, pd.DataFrame([(len(df)+1)*['#']])])
df = df.iloc[:,::-1]
df[len(df)] = '#'
df = df.iloc[::-1,:]
df = pd.concat([df, pd.DataFrame([(len(df)+1)*['#']])])
df = df.rename(columns=dict(zip(df.columns, list(range(len(df))))))
df = df.reset_index(drop=True)
```

```python
point_list = list(zip(np.where(df=='.')[0],np.where(df=='.')[1]))
```

```python
graph = Graph()
for point in tqdm(point_list):
    iterator = iter([0,1,0,-1,1,0,-1,0])
    neighbors = [(point[0]+next(iterator), point[1] + next(iterator)),
                  (point[0]+next(iterator), point[1] + next(iterator)),
                   (point[0]+next(iterator), point[1] + next(iterator)),
                    (point[0]+next(iterator), point[1] + next(iterator))]
    for neighbor in neighbors:
        if neighbor in point_list:
             graph.add_edge(point, neighbor, 1)
```

```python
# GOOD ANSWER : 288
find_path(graph, (1,1), (len(df)-2,len(df)-2))
```
