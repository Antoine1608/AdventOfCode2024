```python
import pandas as pd
import numpy as np
from functools import reduce
from itertools import combinations, chain
```

```python
# get input
input = open("Input.txt").read().split('\n')
input = pd.DataFrame([list(line) for line in input])
input
```

```python
# get the dictionary of antenna coordinates per frequency
list_frequency = list(reduce(set.union, [set(input[i]) for i in range(input.shape[1])]))
list_frequency.remove('.')

antenna_dict = {}
bi_point_combinations = {}
for frequency in list_frequency :
    # get the total combination of bi-points (set of coordinate tuple)
    antenna_dict[frequency] = list(zip(list(np.where(input.values == frequency)[0]),list(np.where(input.values == frequency)[1])))
antenna_dict
```

```python
# for every frequency
binodes = []
for frequency in list_frequency:
    # for every couple of antennas
    for couple in combinations(list(antenna_dict[frequency]), 2):
        # get the couple of binodes
        P1 = couple[0]
        P2 = couple[1]
        node_1 = tuple(np.array(P1) + (np.array(P1) - np.array(P2)))
        node_2 = tuple(np.array(P2) + (np.array(P2) - np.array(P1)))
        binodes.append(node_1)
        binodes.append(node_2)
```

```python
# discard from the list of tuple the tuples out of range and the duplicated tuples 
binodes = set([couple for couple in binodes if ((-1 < couple[0] < input.shape[0]) and (-1 < couple[1] < input.shape[1]))])
```

```python
for couple in binodes:
    input.iloc[couple[0], couple[1]] = '#'
```
